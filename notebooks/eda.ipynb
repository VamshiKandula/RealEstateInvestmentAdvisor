```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"D:\Labmentix\RealEstateInvestmentAdvisor\data\cleaned_housing_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# 1. Distribution of property prices
sns.histplot(df["Price_in_Lakhs"], kde=True)
plt.title("Distribution of Property Prices (Lakhs)")
plt.xlabel("Price (Lakhs)")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["Size_in_SqFt"], kde=True)
plt.title("Distribution of Property Size (SqFt)")
plt.xlabel("Size (SqFt)")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Distribution of property sizes
sns.histplot(df["Size_in_SqFt"], kde=True)
plt.title("Distribution of Property Size (SqFt)")
plt.xlabel("Size (SqFt)")
plt.ylabel("Count")
plt.show()
```

```python
# Outlier detection for Price per SqFt
sns.boxplot(x=df["Price_per_SqFt"])
plt.title("Price per SqFt - Outliers")
plt.show()
```

```python
# Average price by city
df.groupby("City")["Price_in_Lakhs"].mean().sort_values(ascending=False).head(10)
```

```python
# Average Price per SqFt by city
df.groupby("City")["Price_per_SqFt"].mean().sort_values(ascending=False).head(10)
```

```python
# Correlation Analysis
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=["int64", "float64"])
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap (Numeric Only)")
plt.show()
```

```python
# Parking vs Price Analysis
sns.boxplot(x="Parking_Space", y="Price_in_Lakhs", data=df)
plt.title("Parking vs Property Price")
plt.show()
```

```python
# Furnishing Status vs Price Analysis
sns.boxplot(x="Furnished_Status", y="Price_in_Lakhs", data=df)
plt.xticks(rotation=45)
plt.title("Furnishing Status vs Price")
plt.show()
```

```python
# BHK Distribution
sns.countplot(x="BHK", data=df)
plt.title("BHK Distribution")
plt.show()
```

```python
# Size vs Price Analysis
sns.scatterplot(x="Size_in_SqFt", y="Price_in_Lakhs", data=df)
plt.title("Size vs Price")
plt.show()
```

```python
# Nearby Schools vs Price per SqFt Analysis
sns.scatterplot(x="Nearby_Schools", y="Price_per_SqFt", data=df)
plt.title("Nearby Schools vs Price per SqFt")
plt.show()
```
